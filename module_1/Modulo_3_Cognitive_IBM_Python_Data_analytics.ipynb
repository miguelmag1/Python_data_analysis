### GroupBy 

El método groupby se utiliza en variables categóricas, agrupa los datos en subconjuntos según las diferentes categorías de esa variable. Puede agrupar por una sola variable o puede agrupar por múltiples variables.
<br>
<br>
ejemplo: que estamos interesados en encontrar el precio promedio de los vehículos y observar cómo se diferencian entre los diferentes tipos de "estilo de carrocería" y "ruedas motrices", debemos hacer los siguientes pasos:
1. Seleccionamos las tres columnas de datos que nos interesan, lo cual está hecho en esta linea de códogo <code>df_test = df[['drive-wheels','body-style','price']]</code>

2. Agrupamos los datos reducidos según 'ruedas motrices' y 'estilo de carrocería', en el siguiente codigo: <code>df_grp=df_test.groupby(['drive-wheels','body-style'], as_index=False).mean</code>
<br>
Los datos estan agrupados en subcategorias y solo el precio promedio de cada subcategoria se muestra.

### ANOVA (Análisis de varianzas)

ANOVA se puede utilizar para encontrar la correlación entre diferentes grupos de un categórico variable.

De acuerdo con el conjunto de datos del automóvil, podemos usar ANOVA para ver si hay alguna diferencia en la media precio para las diferentes marcas de automóviles como Subaru y Honda.

La prueba ANOVA devuelve dos valores: la puntuación de la prueba F y el valor p.

La prueba F calcula la relación de variación entre la media de los grupos sobre la variación dentro de cada uno de los grupos de muestra. Sin profundizar demasiado en los detalles, la prueba F calcula la relación de variación entre medias de grupo sobre la variación dentro de cada una de las medias de grupo de muestra.

Si la puntuación F es pequeña, la correlación entre el precio como variable objetivo y la agrupaciones es débil, teniendo en cuente que en este caso se evalua un ANOVA entre dos el precio de dos marcas de carros Honda y subaru que no varian fuertemente su diferencia de medias. Asi mismo en el diagrama de barras no se puede ver mucha diferencia en las medias de precio de estas marcas, por ende no se espera un valor F muy alto.

La variación entre los promedios de los dos grupos es comparable a las variaciones dentro de los dos grupos. Suponga que el grupo 1 es "Jaguar" y el grupo 2 es "Honda"; ambos son la función Make categorías. Dado que la puntuación F es grande, la correlación es fuerte en este caso. En este caso entre honda y jaguar se puede esperar un valor F alto pues el diagrama de barras muestra una media mucho menor en precios en honda en comparación con jaguar.

El valor p muestra si el resultado obtenido es estadísticamente significativo.
<br>
<br>
Ejemplo de codigo de ANOVA entre Honda y Subaru.
1. Extraemos los datos de make y precio:
<code>df_anova=df[['make','price']]</code>
2. agrupamos los datos por diferentes marcas:
<code>grouped_anova=df_anova.groupb([['make']])</code>
3. aplicamos la prueba ANOVA usando el paquete scipy, Pasamos los datos de precios de los dos grupos de marcas de automóviles que queremos comparar y calcula los resultados de ANOVA.
<code>anova_results_1=stats.fonewey(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('subaru')['price'])</code>

Los resultados confirman lo que adivinamos al principio. Los precios entre Honda y Subarus no son significativamente diferentes, como la prueba F la puntuación es menor que 1 y el valor p es mayor que 0.05.

podemos hacer los mismo entre honda y jaguar 
<code>anova_results_1=stats.fonewey(grouped_anova.get_group('honda')['price'], grouped_anova.get_group('jaguar')['price'])</code>


Los precios entre Hondas y Jaguars son significativamente diferentes, ya que el puntaje F es muy grande (F = 401) y el valor p es mayor que 0.05.
Con todo, **podemos decir que existe una fuerte correlación entre una variable categórica y otras variables, si la prueba ANOVA nos da un valor de prueba F grande y un valor p pequeño.**

### Correlación de pearson

El método de correlación de Pearson, le dará dos valores, el coeficiente de correlación y el valor p.

Para el coeficiente de correlación, un valor cercano a 1 implica una gran correlación positiva, mientras que un valor cercano a -1 implica una gran correlación negativa y un valor cercano a 0 implica que no hay correlación entre las variables.


A continuación, el valor p nos dirá qué tan seguros estamos de la correlación que calculamos. Para el valor p, un valor menor que 0.001 nos da una fuerte certeza acerca de la correlación coeficiente que calculamos. Un valor entre 0,001 y 0,05 nos da una certeza moderada, un valor entre 0,05 y 0,1 nos dan una certeza débil y un valor p mayor que 0.1 no nos dará certeza de correlación en absoluto.


**Nota**: Podemos decir que existe una fuerte correlación cuando el coeficiente de correlación es cercano a 1 o -1 y el valor p es menor que 0,001.
a continuación veremos una linea de código que permite correlacionar mediante el coeficiente de pearson dos variables
Ejemplo: 
<br>
<code>pearson_coef, p_value = stats.pearson(df['horsepower'], df['priece'])</code>

Podemos ver que el coeficiente de correlación es de aproximadamente 0,8, y esto se acerca a 1 por lo que existe una fuerte correlación positiva. También podemos ver que el valor p es muy pequeño, mucho menor que 0.001 y así podemos concluir que estamos seguros de esta fuerte correlación positiva.
